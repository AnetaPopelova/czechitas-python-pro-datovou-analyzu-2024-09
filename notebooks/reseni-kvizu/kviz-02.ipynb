# Podmíněný výběr tučňáků

## Načtěte dataset Palmer Penguins
Můžete jej stáhnout z GitHub [repozitáře](https://raw.githubusercontent.com/allisonhorst/palmerpenguins/master/inst/extdata/penguins.csv).

Použijte knihovnu pandas, metodu `read_csv()` a data uložte do dataframu nazváného data.

```python
import pandas as pd

url = "https://raw.githubusercontent.com/allisonhorst/palmerpenguins/master/inst/extdata/penguins.csv"

data = pd.read_csv(url)
```

### Které druhy tučňáka mají zastoupení na ostrově Dream?

```python
data[data["island"] == 'Dream']["species"].unique()
```

### Kolik tučňáků má hmotnost vyšší než 5000 gramů?

Použijte podmíněný výběr k filtraci tučňáků s hmotností vyšší než 5000 gramů a spočítejte počet takových záznamů.

```python
data[data["body_mass_g"] > 5_000] #.shape[0]
```

### Kolik tučňáků má hmotnost menší než 3500 gramů a délku křídla (`flipper`) větší než 200 mm?

Použijte podmíněný výběr pro najití tučňáků splňujících obě kritéria a spočítejte, kolik takových záznamů existuje.

```python
data[(data["body_mass_g"] < 3_500) & (data["flipper_length_mm"] > 200)] #.shape[0]
```

### Kolik tučňáků z ostrova Torgensen má ve sloupci `sex` uvedenu hodnotu `female`?

```python
data[(data["island"] == "Torgersen") & (data["sex"] == "female")] #.shape[0]
```

### Který rok má nejvíce záznamů? 

```python
data["year"].value_counts()#.head(1)
```

### Jaká je mediánová hodnota `body_mass_g` tučňáků v datasetu?

Použijte metodu `describe()` pro zobrazení základních statistik.

50\.	percentil je totéž jako medián.

```python
data["body_mass_g"].describe()
```

```python
# Jak se dostat ke konkrétní hodnotě?
# Kód volá metodu describe() na konkrétním sloupci body_mass_g ve dataframe data
# Pomocí indexu ['50%'] vybírá medián (50. percentil)
data["body_mass_g"].describe()['50%']

# Kód volá metodu describe() na celém dataframe data, čímž získá popisné statistiky pro všechny numerické sloupce v dataframe
# Pomocí .loc[] se specifikuje, že chcete hodnotu z řádku '50%' (což odpovídá mediánu) a sloupce 'body_mass_g'
data.describe().loc['50%', 'body_mass_g']
```
