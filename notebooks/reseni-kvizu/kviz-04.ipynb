#### Dokumentace k datasetu “Penguins”

Tento dataset obsahuje údaje o třech druzích tučňáků, kteří se vyskytují v oblasti Palmerovy stanice na Antarktidě. Dataset je často využíván pro výukové účely a analýzy dat.

Dataset obsahuje měření různých fyzických charakteristik tučňáků ze tří různých ostrovů. Data zahrnují informace o délce zobáku, hloubce zobáku, délce ploutví, hmotnosti těla a dalších proměnných. Měření byla provedena na třech druzích tučňáků: Adélie, Gentoo, a Chinstrap.

**Sloupce datasetu:**

- species – Druh tučňáka (Adélie, Gentoo, Chinstrap)
- island – Ostrov, na kterém byl tučňák pozorován (Biscoe, Dream, Torgersen)
- bill_length_mm – Délka zobáku v milimetrech
- bill_depth_mm – Hloubka zobáku v milimetrech
- flipper_length_mm – Délka ploutve v milimetrech
- body_mass_g – Hmotnost těla tučňáka v gramech
- sex – Pohlaví tučňáka (Male nebo Female)
- year – Rok, kdy bylo měření provedeno

````python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt 


url = "https://raw.githubusercontent.com/allisonhorst/palmerpenguins/master/inst/extdata/penguins.csv"

data = pd.read_csv(url)
````

### Jaký je hlavní rozdíl mezi funkcemi `sns.countplot` a `sns.barplot` v Seabornu?

````python
# Obě funkce vytvářejí čárové grafy, ale sns.countplot je používán pro sčítání kategorií v datech, 
# zatímco sns.barplot vykresluje hodnoty na základě souhrnných statistik 
# (např. průměru nebo mediánu) a vyžaduje dodatečné numerické informace.
````

````python
sns.countplot(data, x="species", hue="species")
````

````python
# Vytvoření grafu s barvami a přidání mřížky
ax = sns.countplot(data=data, x="species", hue="species",
                   palette="Set2", legend=False)

# Přizpůsobení os, názvů a stylu
ax.set(
    xlabel="Druh",
    ylabel="Počet",
    title="Počty tučnáků podle druhu",
)

# Přidání gridu pro lepší čitelnost
ax.grid(True, axis='y', linestyle='--', alpha=0.7)

# Nastavení velikosti písma pro osy a titulek
ax.set_xlabel("Druh", fontsize=12)
ax.set_ylabel("Počet", fontsize=12)
ax.set_title("Počty tučnáků podle druhu", fontsize=14)
````

````python
sns.barplot(data=data,
            x="island",
            y="body_mass_g")
````

````python
# Nastavení velikosti grafu
plt.figure(figsize=(10, 6))

# Vytvoření barového grafu pomocí Seaborn s hue pro rozlišení podle druhu
ax = sns.barplot(
    data=data,
    x="island",
    y="body_mass_g",
    palette="pastel",
    hue="island",
    errorbar=None
)

# Přidání popisků na sloupce
for container in ax.containers:
    ax.bar_label(container, fmt="%.0f", label_type="edge", padding=2)

# Nastavení parametrů os a názvu grafu
ax.set(
    xlabel="Ostrov",
    ylabel="Hmotnost (g)",
    title="Průměrná hmotnost tučňáků podle ostrovů a druhů",
)

# Přidání mřížky na osu y pro lepší čitelnost
ax.grid(True, axis='y', linestyle='--', alpha=0.7)

# Nastavení velikosti písma pro osy a titulek
ax.set_xlabel("Ostrov", fontsize=12)
ax.set_ylabel("Průměrná hmotnost (g)", fontsize=12)
ax.set_title("Průměrná hmotnost tučňáků podle ostrovů a druhů", fontsize=14)

# Upravíme legendu, aby byla jasná
ax.legend(title="Druh tučňáků", title_fontsize='13', fontsize='11')
````

### Pro který účel je nejvhodnější použít histogram v Seabornu, například funkci `sns.histplot`?

````python
# Zobrazit rozdělení spojité proměnné

sns.histplot(data["body_mass_g"])
````

### Co ovlivňuje parametr `bins` v funkci sns.histplot?

````python
# Vytvoření histogramu s nastavením počtu binů
ax = sns.histplot(data["body_mass_g"], 
    bins=20, 
    color="peru",
   )

# Nastavení parametrů os a názvu grafu
ax.set(
    xlabel="Hmotnost (g)",
    ylabel="Frekvence",
    title="Histogram hmotnosti tučňáků",
)
````

````python
# Když zadáváte celý dataset (data): je to flexibilnější pro práci s více proměnnými, například při použití hue, col, row, a dalších.
sns.histplot(data=data, x="body_mass_g", hue="species",
             palette="viridis", multiple="stack")
````

### Co znamená, pokud jsou ve boxplotu vidět body mimo hlavní box a whiskers?

````python
# Extrémní hodnoty nebo výjimky

sns.boxplot(data["body_mass_g"])
````

````python
# Vytvoření boxplotu, který rozdělí data podle pohlaví
ax = sns.boxplot(data=data, x="sex", y="body_mass_g", palette="Set2")

# Nastavení parametrů os a názvu grafu
ax.set(
    xlabel="Pohlaví",
    ylabel="Hmotnost (g)",
    title="Rozdělení hmotnosti tučňáků podle pohlaví",
)
````
